```python
# 讀取每個資料並刪除第一行多餘字串
df1 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-06.csv').drop(index=0)
df2 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-07.csv').drop(index=0)
df3 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-08.csv').drop(index=0)
df4 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-09.csv').drop(index=0)
df5 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-10.csv').drop(index=0)
df6 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-11.csv').drop(index=0)
df7 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-12.csv').drop(index=0)
df8 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-13.csv').drop(index=0)
df9 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-14.csv').drop(index=0)
df10 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-15.csv').drop(index=0)
df11 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-16.csv').drop(index=0)
df12 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-17.csv').drop(index=0)
df13 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-18.csv').drop(index=0)
df14 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-19.csv').drop(index=0)
df15 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-20.csv').drop(index=0)
df16 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-21.csv').drop(index=0)
df17 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-22.csv').drop(index=0)
df18 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-23.csv').drop(index=0)
df19 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-24.csv').drop(index=0)
df20 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-25.csv').drop(index=0)
df21 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-26.csv').drop(index=0)
df22 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-27.csv').drop(index=0)
df23 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-02-28.csv').drop(index=0)
df24 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-01.csv').drop(index=0)
df25 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-02.csv').drop(index=0)
df26 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-03.csv').drop(index=0)
df27 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-04.csv').drop(index=0)
df28 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-05.csv').drop(index=0)
df29 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-06.csv').drop(index=0)
df30 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-07.csv').drop(index=0)
df31 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-08.csv').drop(index=0)
df32 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-09.csv').drop(index=0)
df33 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-10.csv').drop(index=0)
df34 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-11.csv').drop(index=0)
df35 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-12.csv').drop(index=0)
df36 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-13.csv').drop(index=0)
df37 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-14.csv').drop(index=0)
df38 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-15.csv').drop(index=0)
df39 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-16.csv').drop(index=0)
df40 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-17.csv').drop(index=0)
df41 = pd.read_csv('/content/MyGoogleDrive/MyDrive/統整資料/467420-2023-03-18.csv').drop(index=0)

# 將多個DataFrame合併
merged_df = pd.concat([df1, df2 ,df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20, df21, df22 ,df23, df24, df25, df26, df27, df28, df29, df30, df31, df32 ,df33, df34, df35, df36, df37, df38, df39, df40, df41])

# 檢查有多少遺失值
merged_df.isna().sum()

# 重新設定索引及刪除重複值
merged_df = merged_df.reset_index(drop=True).drop_duplicates()

# 將部分欄位的資料類型轉換為float型態
merged_df['風速(m/s)'] = merged_df['風速(m/s)'].astype(float)
merged_df['氣溫(℃)'] = merged_df['氣溫(℃)'].astype(float)
merged_df['露點溫度(℃)'] = merged_df['露點溫度(℃)'].astype(float)

# 計算熱量欄位
merged_df['熱量(cal)'] = merged_df['風速(m/s)'] * 0.086 * (merged_df['氣溫(℃)'] - merged_df['露點溫度(℃)']) * 10

# 繪製日期
start_date = '2023-02-06'
end_date = '2023-03-18'
date_range = pd.date_range(start=start_date, end=end_date)
date_range = date_range.repeat(24)

merged_df['日期(date)'] = pd.Series(date_range)

# 將日期轉換為日期時間物件
date = pd.to_datetime(merged_df['日期(date)'], format='%Y/%m/%d')

# 將 '觀測時間(hour)' 與日期時間物件相加，得到新的 'step' 欄位
merged_df['觀測時間(hour)'] = merged_df['觀測時間(hour)'].astype(int)
merged_df['日期+時數(step)'] = date + pd.to_timedelta(merged_df['觀測時間(hour)'], unit='h')

# 選擇 '氣溫(℃)' 和 '熱量(cal)' 兩個欄位
merged_df['氣溫(℃)'] = merged_df['氣溫(℃)'].astype(float)
merged_df['熱量(cal)'] = merged_df['熱量(cal)'].astype(float)
df_X = merged_df[['氣溫(℃)', '熱量(cal)']]

# 資料標準化
scaler = StandardScaler()
df_X_scaled = scaler.fit_transform(df_X)

# 設定 K-means 模型參數
kmeans = KMeans(n_clusters=3).fit(df_X_scaled)
pred = kmeans.predict(df_X_scaled)

# 將預測結果加入原始資料
df1 = pd.DataFrame(df_X, columns=df_X.columns)
df1['pred'] = pred

# 繪製分群結果
c = {0: 'r', 1:'g', 2:'b'}
df1['colors'] = df1['pred'].map(c)
df1.plot(kind='scatter', x='氣溫(℃)', y='熱量(cal)', c=df1['colors'])

# 將合併後的資料儲存為CSV檔案
merged_df.to_csv('/content/MyGoogleDrive/MyDrive/統整資料/彙整資料集_110B16890_顏辰霖.csv', index=False)
```
